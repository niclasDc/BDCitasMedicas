```python
from pandasql import sqldf
import pandas as pd
mysql = lambda q: sqldf(q, globals())
```

```python
direccion_url = 'https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)'
datos_url_df = pd.read_html(direccion_url)
datos_countries_df = datos_url_df[0]
datos_countries_df = datos_countries_df[datos_countries_df.columns]
datos_countries_df.columns = ['Country','ContinentalRegion','StatisticalSubregion','Population2018','Population2019','Change']
datos_countries_df
```

1. Pais con mayor población durante 2019

```python
query = '''SELECT Country, ContinentalRegion, StatisticalSubregion 
                FROM (SELECT *, Max(Population2019) as MayorPoblacion 
                            FROM datos_countries_df WHERE Country != 'World' );'''
mysql(query)
```

2. Pais con mayor tasa de crecimiento para el periodo 2018-2019

```python
query = '''SELECT Country, ContinentalRegion, StatisticalSubregion 
                FROM (SELECT * , MAX(((CAST(Population2019 AS float)/CAST(Population2018 AS float))-1)*100) 
                            FROM datos_countries_df);'''
mysql(query)
```

3. Pais con menor tasa de crecimiento para el periodo 2018-2019

```python
query = '''SELECT Country, ContinentalRegion, StatisticalSubregion 
                FROM (SELECT *, MIN(((CAST(Population2019 AS float)/CAST(Population2018 AS float))-1)*100) 
                            FROM datos_countries_df);'''
mysql(query)
```

4. Tasa de crecimiento de Asia para el periodo 2018-2019

```python
query = '''SELECT ContinentalRegion, ((CAST(SUM(Population2019) AS float)/CAST(SUM(Population2018) AS float))-1)*100 AS TasaCrecimientoAsia2018_2019 
                FROM datos_countries_df 
                WHERE ContinentalRegion = 'Asia';'''
mysql(query)
```

5. Tasa de crecimiento de America del sur para el periodo 2018-2019

```python
query = '''SELECT ContinentalRegion, StatisticalSubregion, ((CAST(SUM(Population2019) AS float)/CAST(SUM(Population2018) AS float))-1)*100 AS TasaCrecimientoAmericaSur2018_2019 
                FROM datos_countries_df 
                WHERE ContinentalRegion = 'Americas'  
                AND StatisticalSubregion = 'South America';'''
mysql(query)
```

6. Los continentes con mayor y menor tasa de crecimiento, respectivamente

```python
query = '''SELECT ContinentalRegion, ((CAST(SUM(Population2019) AS float)/CAST(SUM(Population2018) AS float))-1)*100 AS TasaCrecimientoContinente2018_2019 
                FROM datos_countries_df 
                WHERE ContinentalRegion != 'None' 
                GROUP BY ContinentalRegion;'''
mysql(query)
```
